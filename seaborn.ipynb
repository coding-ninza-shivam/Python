# Seaborn

This is library for creating statistical visalization.

### Categorise of plots in seaborn

1. Relatonial plot : (linear plot, scatter plot)
2. Catogorical plot : (box plot. volini plot, bar plot)
3. Distribution plot : (histplot, kdeplot)
4. Regressoin plot : (lmplot, reglpot)
5. Matrix plot : (heatmap)
6. Multi-plot grid : (FacetGrid)

# Basic plots in seaborn

### 1. Lineplot

```python
import seaborn as sns
import matplotlib.pyplot as plt

fmri = sns.load_dataset("fmri")
sns.lineplot(x="timepoint", y="signal", hue="region", data=fmri)
plt.show()
```

### 2. Scatter plot

```python
tips = sns.load_dataset("tips")

sns.scatterplot(x="total_bill", y="tip", hue="day", data=tips)
plt.show()
```

### 3.Barplot

```python
sns.barplot(x="day", y="total_bill", data=tips)
plt.show()
```

### 4. Boxplot

```python
sns.boxplot(x="day", y="total_bill", data=tips)
plt.show()
```

### 5. Histplot

```python
sns.histplot(tips["total_bill"], kde=True)
plt.show()
```

### 6. Heatmap

```python
corr = tips.corr(numeric_only = True)
 
sns.heatmap(corr, annot=True, cmap = "coolwarm")
plt.show()
```

### 7. Pairplot

```python
iris = sns.load_dataset("iris")
sns.pairplot(iris, hue="species")
plt.show()
```
