# Panda Series
One dimensional array holding data of anytype

```python
import pandas as pd
a = [1, 7, 2]
myvar = pd.Series(a)
print(myvar)
```

```python
# Creating our own label
import pandas as pd
a = [1, 7, 2]
myvar = pd.Series(a, index = ["x", "y", "z"])
print(myvar)
print(myvar["y"])
```

```python
# Using key value objects
calories = {"day1": 420, "day2": 380, "day3": 390}
myvar = pd.Series(calories)
print(myvar)
```

# DataFrame

```python
data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}
df = pd.DataFrame(data)
print(df)
```

### Locate Row

```python
print(df.loc[0])

print(df.loc[[0, 1]])


df = pd.DataFrame(data, index = ["day1", "day2", "day3"])
#refer to the named index:
print(df.loc["day2"])
```

# Cleaning Data

## Cleaning empty cells

```python
new_df = df.dropna()  # Will not change the original dataframe
print(new_df.to_string())
```

```python
new_df = df.dropna(inplace=True)  # Will  change the original dataframe
```

Replace empty values in spite of deleting empty rows

```python
df = pd.read_csv('data.csv')
df.fillna(130, inplace = True)
```

```python
df = pd.read_csv('data.csv')
df.fillna({"Calories": 130}, inplace=True)
```

Replce using mean, meadian and mode

```python
df = pd.read_csv('data.csv')
df.fillna({"Calories": df["Calories"].mean()}, inplace=True)
```

```python
df = pd.read_csv('data.csv')
df.fillna({"Calories": df["Calories"].median()}, inplace=True)
```

```python
df = pd.read_csv('data.csv')
df.fillna({"Calories": df["Calories"].mode()[0]}, inplace=True)
```

## Cleaning wrong format

```python
# Convert into correct format
df = pd.read_csv('data2.csv')
df['Date'] = pd.to_datetime(df['Date'], format='mixed')
print(df.to_string())
```

```python
# Removing rows
df.dropna(subset=['Date'], inplace = True)
```

## cleaning wrong data

```python
df = pd.read_csv('data3.csv')

df.drop(df[df["Duration"] > 120].index, inplace=True)
```

## Cleaning duplicates

```python
df = pd.read_csv('data3.csv')

df.drop_duplicates(inplace = True)
```
